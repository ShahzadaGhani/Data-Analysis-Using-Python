```python
# use groupby() to generate sum,we group data on the Year column and calculate the sum for all numeric columns of DataFrame, and also sort Year in descending order.
df.groupby(['Year'],as_index=False)['Revenue'].sum().sort_values(by='Revenue',ascending=False)
```

```python
df.groupby(['Year'],as_index=False)['Revenue'].sum().sort_values(by='Revenue',ascending=False)
# plotting a bar plot for year column
sns.barplot(x='Year',y='Revenue',data=df)
```

```python
#total no of tourist from top 4 country
Top4=df.groupby(['Country'],as_index=False)['Number_of_Tourists'].sum().sort_values(by='Number_of_Tourists',ascending=False).head(4)

sns.barplot(data=Top4 ,x='Country',y='Number_of_Tourists',palette='viridis')#,data=df)
```

```python
df.groupby(['Number_of_Tourists'],as_index=False)['Revenue'].sum().sort_values(by='Revenue',ascending=False).head(7)
```

```python
#max_Revenue 
df.groupby(['Number_of_Tourists'],as_index=False)['Revenue'].sum().sort_values(by='Revenue',ascending=False).head(5)
sns.set(rc={'figure.figsize':(20,5)})
sns.barplot(x='Revenue',y='Number_of_Tourists',data=df)
```

```python
#countplot() method is used to Show the counts of observations in each categorical bin using bars.
#count plot on single categorical variable
sns.countplot(data=df, x ='Number_of_Tourists')
sns.set(rc={'figure.figsize':(25,3)})
plt.title('OverTime')
plt.show()
```

```python
#count plot on two categorical variable
# hue is use to charcterized and differentiate but not for multiple attribute
ax=sns.countplot(x='Revenue',hue = 'Number_of_Tourists',data=df)
sns.set(rc={'figure.figsize':(21,3)})
# for assigning values 
for bars in ax.containers:
    ax.bar_label(bars)
```

```python
# relationship between Number_of_Tourists and Revenue over time
plt.figure(figsize=(10, 6))
sns.lineplot(x='Year', y='Number_of_Tourists', data=df, label='Tourist Count')
sns.lineplot(x='Year', y='Revenue', data=df, label='Revenue')
plt.xticks(rotation=45)
plt.title('Tourist Count and Revenue Over Time')
plt.legend()
plt.show()
```

```python
# Revenue distribution with the help of histplot
sns.histplot(df['Revenue'], bins=19 , kde=False, color='blue')
plt.title('Revenue Distribution')
plt.xlabel('Revenue')
plt.ylabel('Frequency')
plt.show()
```
